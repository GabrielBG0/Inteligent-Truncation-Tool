```python
import pandas as pd
ds = pd.read_csv('../../CSVs/dataset_predictionsInt.csv', sep='ยง')
ds
```

```python
for index, row in ds.iterrows():
    score = 0
    if row['label'] == row['predictionsMLPC']:
        score += 1
    if row['label'] == row['predictionsSVC']:
        score += 1
    if row['label'] == row['predictionsRFC']:
        score += 1
    if row['label'] != row['predictionsMLPC']:
        score -= 1
    if row['label'] != row['predictionsSVC']:
        score -= 1
    if row['label'] != row['predictionsRFC']:
        score -= 1
    ds.loc[index, 'score'] = score

ds
```

```python
ds.to_csv('../../CSVs/dataset_predictionsIntScore.csv',
          sep='ยง', index=False, columns=['text', 'score'])
```
